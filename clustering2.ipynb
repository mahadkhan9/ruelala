```python
df=pd.read_csv("ass4q2.csv")
df
```

Training data for clustering with Non-Sold out items only

```python
x = df[df["Sold Out"] == "No"]
x
```

```python
j = x.iloc[:, 2:-2]
j
```

**For K = 2**

```python
ks = [2]
for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=10, n_init="auto").fit(j)
```

```python
a = kmeans.cluster_centers_
a
```

```python
z = pd.DataFrame(a, columns=range(1, 25))
z
```

**For K = 4**

```python
kt = [4]
for i in kt:
    kmeans = KMeans(n_clusters=i, random_state=10, n_init="auto").fit(j)
```
